*APPLY BFS AND DFS ON TREES AND GRAPHS*

```python
""" Graph
A — B — D
|   |
C   E
|
F
```

**BREADTH FIRST SEACH IMPLEMENTATION**

---



```python
from collections import deque

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

def bfs(start):
    visited = []
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            print(node, end=" ")
            visited.append(node)
            queue.extend(graph[node])

bfs('A')
```

**DEAPTH FIRST SEARCH IMPLENTATION**

---







```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

def dfs(node, visited):
    if node not in visited:
        print(node, end=" ")
        visited.append(node)
        for neighbour in graph[node]:
            dfs(neighbour, visited)

visited = []
dfs('A', visited)
```

```python
"""TREE
        A
      /   \
     B     C
    / \     \
   D   E     F
```

**BREADTH FIRST SEARCH IMPLEMENTATION**

---



```python
from collections import deque  # for queue

# tree as graph
tree = {
    'A': ['B', 'C'],  # root node
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

def bfs(start):
    visited = []              # track visited nodes
    queue = deque([start])    # queue for BFS

    while queue:
        node = queue.popleft()   # remove from front
        if node not in visited:
            print(node, end=" ")
            visited.append(node)
            queue.extend(tree[node])  # add children

bfs('A')
```

**DEAPTH FIRST SEARCH IMPLEMENTATION**

---



```python
# same tree
tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

def dfs(node, visited):
    if node not in visited:
        print(node, end=" ")
        visited.append(node)

        for child in tree[node]:   # go deep
            dfs(child, visited)

visited = []
dfs('A', visited)
```

# ***LAB TASK NO 2***

```python
import random
import time
import pandas as pd
import matplotlib.pyplot as plt
from collections import deque

# Step 1: create datasets
sizes = [1000, 40000, 80000, 200000, 1000000]
results = []

# BFS (unchanged - already correct)
def bfs(lst, goal):
    visited = set()
    queue = deque([0])  # start from index 0

    while queue:
        i = queue.popleft()

        if i in visited:
            continue
        visited.add(i)

        if lst[i] == goal:
            return True

        if i + 1 < len(lst):
            queue.append(i + 1)


# FIXED DFS
def dfs(lst, goal):
    visited = set()
    stack = [0]  # start from index 0

    while stack:
        i = stack.pop()

        if i in visited:
            continue
        visited.add(i)

        if lst[i] == goal:
            return True

        if i + 1 < len(lst):
            stack.append(i + 1)


# Step 2–3: run for each dataset
for size in sizes:
    print(f"Processing size: {size}")

    # generate unique random numbers
    lst = random.sample(range(size * 10), size)

    # goal = last - 220 index
    goal = lst[len(lst) - 220]

    # BFS timing
    start = time.time()
    bfs(lst, goal)
    bfs_time = time.time() - start

    # DFS timing
    start = time.time()
    dfs(lst, goal)
    dfs_time = time.time() - start

    # store results
    results.append([size, bfs_time, dfs_time])


# Step 4: create dataframe
df = pd.DataFrame(results, columns=["Size", "BFS Time", "DFS Time"])
print(df)

# Step 5: plot bar chart
df.plot(x="Size", y=["BFS Time", "DFS Time"], kind="bar")

plt.xlabel("Input Size")
plt.ylabel("Time (seconds)")
plt.title("BFS vs DFS Time Comparison")

plt.tight_layout()  # fixes display issues
plt.show()
```

# ***lab task 3***

```python
from collections import deque

# Graph representation
graph = {
    "Islamabad": ["Rawalpindi", "Lahore", "Peshawar"],
    "Rawalpindi": ["Islamabad", "Peshawar", "Quetta"],
    "Peshawar": ["Islamabad", "Rawalpindi", "Quetta"],
    "Lahore": ["Islamabad", "Multan", "Quetta"],
    "Multan": ["Lahore", "Karachi", "Quetta"],
    "Quetta": ["Rawalpindi", "Peshawar", "Multan", "Karachi"],
    "Karachi": ["Multan", "Quetta"]
}

def bfs_shortest_path(start, goal):
    visited = set()
    queue = deque([[start]])  # store paths

    while queue:
        path = queue.popleft()
        city = path[-1]

        if city == goal:
            return path

        if city not in visited:
            visited.add(city)

            for neighbor in graph[city]:
                new_path = list(path)
                new_path.append(neighbor)
                queue.append(new_path)

# find path
start = "Islamabad"
goal = "Karachi"

path = bfs_shortest_path(start, goal)

print("Shortest Path:", " → ".join(path))
print("Number of cities:", len(path))
```
