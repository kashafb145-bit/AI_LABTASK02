# *Activity number 1*

```python
graph = {
    '5': ['3', '7'],
    '3': ['2', '4'],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}

visited = []  # List for visited nodes
queue = []    # Initialize a queue

def bfs(visited, graph, node):  # function for BFS
    visited.append(node)
    queue.append(node)

    while queue:  # Creating loop to visit each node
        m = queue.pop(0)
        print(m, end=" ")

        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)

# Driver Code
print("Following is the Breadth-First Search")
bfs(visited, graph, '5')  # function calling
```

# *Activity number2*

```python
def printParents(node, adj, parent):
    # current node is Root, thus, has no parent
    if parent == 0:
        print(node, "-> Root")
    else:
        print(node, "->", parent)

    # Using DFS
    for cur in adj[node]:
        if cur != parent:
            printParents(cur, adj, node)


# Function to print the children of each node
def printChildren(Root, adj):
    # Queue for BFS
    q = []

    # pushing the root
    q.append(Root)

    # visit array
    vis = [0] * len(adj)

    while len(q) > 0:
        node = q.pop(0)
        vis[node] = 1

        print(node, "-> ", end=" ")

        for cur in adj[node]:
            if vis[cur] == 0:
                print(cur, end=" ")
                q.append(cur)

        print("\n")


# Function to print the leaf nodes
def printLeafNodes(Root, adj):
    # Leaf nodes have only one edge and are not the root
    for i in range(1, len(adj)):
        if len(adj[i]) == 1 and i != Root:
            print(i, end=" ")
    print("\n")


# Function to print degrees of each node
def printDegrees(Root, adj):
    for i in range(1, len(adj)):
        print(i, ": ", end=" ")

        if i == Root:
            print(len(adj[i]))
        else:
            print(len(adj[i]))


# Driver code
N = 7
Root = 1

# Adjacency list
adj = []
for i in range(N + 1):
    adj.append([])

# Creating the tree
adj[1].append(2)
adj[2].append(1)

adj[1].append(3)
adj[3].append(1)

adj[1].append(4)
adj[4].append(1)

adj[2].append(5)
adj[5].append(2)

adj[2].append(6)
adj[6].append(2)

adj[4].append(7)
adj[7].append(4)

# Output
print("The parents of each node are:")
printParents(Root, adj, 0)

print("\nThe children of each node are:")
printChildren(Root, adj)

print("The leaf nodes of the tree are:")
printLeafNodes(Root, adj)

print("The degrees of each node are:")
printDegrees(Root, adj)
```

```python
from collections import deque

# Weighted Graph (node: [(neighbor, weight)])
graph = {
    'A': [('B', 2), ('C', 1)],
    'B': [('D', 4)],
    'C': [('D', 2), ('E', 3)],
    'D': [('G', 1)],
    'E': [('G', 5)],
    'G': []
}

# BFS with cost calculation
def bfs_with_cost(graph, start, goal):
    queue = deque()
    queue.append((start, 0, [start]))  # (node, cost, path)

    visited = []

    while queue:
        node, cost, path = queue.popleft()

        if node not in visited:
            visited.append(node)

            # Goal check
            if node == goal:
                print("Goal Reached!")
                print("Path:", " -> ".join(path))
                print("Total Cost:", cost)
                return

            # Explore neighbors
            for neighbour, weight in graph[node]:
                queue.append((neighbour, cost + weight, path + [neighbour]))

    print("Goal not found!")


# Driver Code
print("=== BFS with Cost Calculation ===")

start_node = 'A'
goal_node = 'G'

bfs_with_cost(graph, start_node, goal_node)
```
