# ***ucs implementaton ***

python code to implement the ucs **algorithm**

---



```python
import heapq

# Graph representation (adjacency list with costs)
graph = {
    'S': [('A', 3), ('B', 2), ('C', 7)],
    'A': [('D', 3), ('E', 8), ('G', 15)],
    'B': [('G', 20)],
    'C': [('G', 6)],
    'D': [],
    'E': [],
    'G': []
}

def uniform_cost_search(start, goal):
    # priority queue: (cost, node, path)
    pq = []
    heapq.heappush(pq, (0, start, [start]))

    # visited dictionary to store best cost
    visited = {}

    while pq:
        cost, node, path = heapq.heappop(pq)

        # If goal is reached
        if node == goal:
            return cost, path

        # If we already visited with a lower cost, skip
        if node in visited and visited[node] <= cost:
            continue

        visited[node] = cost

        # Expand neighbors
        for neighbor, edge_cost in graph[node]:
            new_cost = cost + edge_cost
            new_path = path + [neighbor]
            heapq.heappush(pq, (new_cost, neighbor, new_path))

    return float("inf"), []

# Run UCS
cost, path = uniform_cost_search('S', 'G')

print("Minimum Cost:", cost)
print("Optimal Path:", " -> ".join(path))
```
