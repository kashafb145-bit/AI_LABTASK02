display all attributes and their values.
*italicised text*

```python
class Student:
    def __init__(self, student_id, name, student_class):
        self.student_id = student_id
        self.name = name
        self.student_class = student_class

    def display(self):
        for key, value in self.__dict__.items():
            print(key, ":", value)

s3 = Student(2, "Zara", "BSAI")
s3.display()
```

# ***Lab Task 2: Object-Oriented Design and Area Comparison***


*Design two classes representing Trapezoid and Parallelogram. Define appropriate attributes and compute their areas using standard mathematical formulas.*

```python
# Trapezoid class
class Trapezoid:
    def __init__(self, a, b, height):
        self.a = a
        self.b = b
        self.height = height

    def area(self):
        return 0.5 * (self.a + self.b) * self.height


# Parallelogram class
class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height


# Store areas in dictionary
areas = {}

# Multiple inputs
shapes = [
    ("Trapezoid1", Trapezoid(5, 7, 4)),
    ("Parallelogram1", Parallelogram(6, 5)),
    ("Trapezoid2", Trapezoid(8, 10, 3)),
    ("Parallelogram2", Parallelogram(9, 6))
]

# Calculate and store areas
for name, shape in shapes:
    areas[name] = shape.area()

# Display areas
print("\nAreas of shapes:")
for name, area in areas.items():
    print(name, ":", area)

# Find maximum area
max_shape = max(areas, key=areas.get)

print("\nShape with maximum area:")
print(max_shape, "with area =", areas[max_shape])
```
